# Import Libraries  

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Import Data

```python
bottle = pd.read_csv('bottle.csv', low_memory=False)
pd.set_option('display.float_format', lambda x:'%f'%x)
```

# Data management 

```python
bottle['R_Depth'] = pd.to_numeric(bottle['R_Depth'], errors='coerce')
```

```python
bottle['R_TEMP'] = pd.to_numeric(bottle['R_TEMP'], errors='coerce')
```

```python
bottle['Cst_Cnt'] = pd.to_numeric(bottle['Cst_Cnt'], errors='coerce')
```

```python
bottle['R_SALINITY'] = pd.to_numeric(bottle['R_SALINITY'], errors='coerce')
```

# Box plot

# Histogram 

```python
fig = plt.figure()
sub1 = bottle['R_Depth']
sns.distplot(sub1, kde=False);
plt.title('Approximate number of reports taken at different depths for bottle dataset')
plt.xlabel('Depth')
```

# Line chart 

```python
fig = plt.figure()
sub2 = bottle[(bottle['Cst_Cnt']==1)]
var = sub2['R_SALINITY'].dropna()
%matplotlib notebook           
var.plot(kind='line')
plt.title('Salinity readings for Cast 1')
plt.ylabel('Salinity')
```
